```c++
TFile DATA("result.root");
TFile TT("TTJetsDiLep.root");  
TFile MuEG("MuEG.root");
```

```c++
%jsroot on
```

```c++
TH1F* h_cosMuMuData = (TH1F*) DATA.Get("Whelicity/h_cosMuMu");
TH1F* h_cosMuMuTT = (TH1F*) TT.Get("Whelicity/h_cosMuMu");
TH1F* h_cosMuElTT = (TH1F*) TT.Get("Whelicity/h_cosElMu");
TH1F* h_cosMuElData = (TH1F*) MuEG.Get("Whelicity/h_cosElMu");
```

```c++
gStyle->SetOptStat(1111);
// signal TTbar
h_cosMuMuTT->SetFillStyle(1001);
h_cosMuMuTT->SetFillColor(kRed);
h_cosMuMuTT->SetLineColor(kRed);
h_cosMuMuTT->Scale(1/h_cosMuMuTT->Integral("width"));
h_cosMuMuTT->SetTitle("TT");
h_cosMuMuTT->Draw("Hist");

// Data 
h_cosMuMuData->Scale(1/h_cosMuMuData->Integral("width"));
h_cosMuMuData->SetLineColor(1);
h_cosMuMuData->SetMarkerStyle(10);
h_cosMuMuData->SetLineStyle(6);
h_cosMuMuData->SetTitle("DATA");
h_cosMuMuData->Draw("E1 same");

leg = new TLegend(0.1+.6,0.7,0.48+.4,0.9);
// leg->SetHeader("The Legend Title","C"); // option "C" allows to center the header
leg->AddEntry(h_cosMuMuData,"Data MuMu");
leg->AddEntry(h_cosMuMuTT,"TT signal");
leg->Draw();

c1->Draw();
c1->SaveAs("MuMu.pdf");



```

```c++
TH1F* RatioTop =(TH1F*) h_cosMuMuData->Clone();
TH1F* RatioBottom =(TH1F*) h_cosMuMuTT->Clone();
RatioTop->Divide(RatioBottom);
```

```c++
RatioTop->SetTitle("");
RatioTop->GetYaxis()->SetTitle("DATA/MC");
RatioTop->SetTitleSize(0.1, "XYZ");
RatioTop->SetTitleOffset(0.55, "X");
RatioTop->SetTitleOffset(0.3, "Y");
RatioTop->SetLabelSize(0.06,"XY");
RatioTop->GetYaxis()->SetRangeUser(0.4,1.6);
RatioTop->Draw();
c1->Draw();
```

```c++
TF1* FitFunc = new TF1("fitfunc","[0]*(3./8)*(1-x)^2+(1-[0]-[1])*(3./8)*(1+x)^2+(3./4)*(1-x^2)*[1]",-1,1);
```

```c++
h_cosMuMuTT->Fit(FitFunc);
```

```c++
h_cosMuMuData->Fit(FitFunc);
```

```c++

h_cosMuMuData->Draw();
h_cosMuMuTT->SetFillColorAlpha(kRed,.5);
h_cosMuMuTT->Draw("hist same");

c1->Draw();
```

```c++
// signal TTbar
h_cosMuElTT->SetFillStyle(1001);
h_cosMuElTT->SetFillColor(kRed);
h_cosMuElTT->SetLineColor(kRed);
h_cosMuElTT->Scale(1/h_cosMuElTT->Integral("width"));
h_cosMuElTT->SetTitle("TT");
h_cosMuElTT->Draw("Hist");

// Data 
h_cosMuElData->Scale(1/h_cosMuElData->Integral("width"));
h_cosMuElData->SetLineColor(1);
h_cosMuElData->SetMarkerStyle(10);
h_cosMuElData->SetLineStyle(6);
h_cosMuElData->SetTitle("DATA");
h_cosMuElData->Draw("E1 same");
leg = new TLegend(0.1+.6,0.7,0.48+.4,0.9);
// leg->SetHeader("The Legend Title","C"); // option "C" allows to center the header
leg->AddEntry(h_cosMuElData,"Data ElMu");
leg->AddEntry(h_cosMuElTT,"TT signal");
leg->Draw();
c1->Draw();
c1->SaveAs("MuEl.pdf");
```

```c++
h_cosMuElData->Fit(FitFunc);
h_cosMuElTT->Fit(FitFunc);
```

```c++
h_cosMuElData->Draw();
h_cosMuElTT->SetFillColorAlpha(kRed,.5);
h_cosMuElTT->Draw("hist same");

c1->Draw();
```
